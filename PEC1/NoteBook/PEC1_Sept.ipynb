# Machine Learning
To facilitate the work with the algorithms we are going to use the dataset from sklearn that has the class data in numerical form

```python
# load iris
iris_sk = datasets.load_iris()
```

```python
# separation of the data into training and test to test the algorithms
X_train, X_test, y_train, y_test = train_test_split(iris_sk.data, iris_sk.target, test_size=0.25, random_state=4)
print('Son {} datos para entrenamiento y {} datos para prueba'.format(X_train.shape[0], X_test.shape[0]))
```

# Classification: Training
As it is a classification problem, we will use the main classifiers

1.   Logistic Regression 
2.   K-Nearest Neighbors (K-NN)
1.   Support Vector Machine (SVM)
2.   Naive Bayes
1.   Decision Tree Classification
2.   Random Forest Classification

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# variable scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

image.png

```python
# fit the model: Logistic Regression
classifierLR = LogisticRegression(random_state = 0)
classifierLR.fit(X_train, y_train)
# k-fold cross validation
accuraciesLR = cross_val_score(estimator = classifierLR, X = X_train, y = y_train, cv = 10)
accuraciesLR.mean()
accuraciesLR.std()
out = " %f (%f)" % (accuraciesLR.mean(), accuraciesLR.std())
print(out)
```

image.png

```python
# fit the model: KNN 
classifierKNN = KNeighborsClassifier(n_neighbors = 5, metric = "minkowski", p = 2)
classifierKNN.fit(X_train, y_train)
# k-fold cross validation
accuraciesKNN = cross_val_score(estimator = classifierKNN, X = X_train, y = y_train, cv = 10)
accuraciesKNN.mean()
accuraciesKNN.std()
out = " %f (%f)" % (accuraciesKNN.mean(), accuraciesKNN.std())
print(out)
```

image.png

```python
# fit the model: SVM 
classifierSVM = SVC()
classifierSVM.fit(X_train, y_train)
# k-fold cross validation
accuraciesSVM = cross_val_score(estimator = classifierSVM, X = X_train, y = y_train, cv = 10)
accuraciesSVM.mean()
accuraciesSVM.std()
out = " %f (%f)" % (accuraciesSVM.mean(), accuraciesSVM.std())
print(out)
```

image.png

```python
# fit the model: Naive Bayes
classifierGNB = GaussianNB()
classifierGNB.fit(X_train, y_train)
# k-fold cross validation
accuraciesGNB = cross_val_score(estimator = classifierGNB, X = X_train, y = y_train, cv = 10)
accuraciesGNB.mean()
accuraciesGNB.std()
out = " %f (%f)" % (accuraciesGNB.mean(), accuraciesGNB.std())
print(out)
```

image.png

```python
# fit the model: Decision Tree
classifierDT = DecisionTreeClassifier(criterion = "entropy", random_state = 0)
classifierDT.fit(X_train, y_train)
# k-fold cross validation
accuraciesDT = cross_val_score(estimator = classifierDT, X = X_train, y = y_train, cv = 10)
accuraciesDT.mean()
accuraciesDT.std()
out = " %f (%f)" % (accuraciesDT.mean(), accuraciesDT.std())
print(out)
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize = (10,7))
plot_tree(classifierDT.fit(X_train, y_train)  ,filled=True)
plt.show()
```

image.png

```python
# fit the model: Random Forest
classifierRF = RandomForestClassifier(n_estimators = 10, criterion = "entropy", random_state = 0)
classifierRF.fit(X_train, y_train)
# k-fold cross validation
accuraciesRF = cross_val_score(estimator = classifierRF, X = X_train, y = y_train, cv = 10)
accuraciesRF.mean()
accuraciesRF.std()
out = " %f (%f)" % (accuraciesRF.mean(), accuraciesRF.std())
print(out)
```

# Clasification: Prediction

```python
# Prediction of the results with the Test Set
y_pred  = classifierDT.predict(X_test)

# confusion matrix
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print('Matriz de Confusin')
print(cm)
print('Precisin rbol de Decisin: {}'.format(classifierDT.score(X_train, y_train)))
print('Reporte de Clasificacin')
print(classification_report(y_test, y_pred))
```

# 5x2cv Model Clasification

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

from mlxtend.data import iris_data
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

X, y = iris_data()

c1 = LogisticRegression(random_state=1, solver='liblinear', multi_class='ovr')
c2 = GaussianNB()
c3 = KNeighborsClassifier(n_neighbors = 5, metric = "minkowski", p = 2)

X_train, X_test, y_train, y_test = \
    train_test_split(X, y, test_size=0.25,random_state=123)

score1 = c1.fit(X_train, y_train).score(X_test, y_test)
score2 = c2.fit(X_train, y_train).score(X_test, y_test)
score3 = c3.fit(X_train, y_train).score(X_test, y_test)

y_target = y_test

y_model1= c1.fit(X_train, y_train).predict(X_test)
y_model2= c2.fit(X_train, y_train).predict(X_test)
y_model3= c3.fit(X_train, y_train).predict(X_test)


#classification report
cr1= classification_report(y_test, y_model1)
cr2= classification_report(y_test, y_model2)
cr3 = classification_report(y_test, y_model3)
# confusion matrix
cm1 = confusion_matrix(y_test, y_model1)
cm2 = confusion_matrix(y_test, y_model2)
cm3 = confusion_matrix(y_test, y_model3)
# root mean square error
rms1 = mean_squared_error(y_test, y_model1, squared=False)
rms2 = mean_squared_error(y_test, y_model2, squared=False)
rms3 = mean_squared_error(y_test, y_model3, squared=False)

print('Logistic regression classification report: ')
print(cr1)
print('NB Gauss classification report: ')
print(cr2)
print('Knn classification report: ')
print(cr3)
print('Logistic regression accuracy: %.2f%%' % (score1*100))
print(cm1)
print('RMS', rms1)
print('NB Gauss accuracy: %.2f%%' % (score2*100))
print(cm2)
print('RMS', rms2)
print('Knn accuracy: %.2f%%' % (score3*100))
print(cm3)
print('RMS', rms3)

```

# Paired Test 5x2cv

```python
from mlxtend.evaluate import paired_ttest_5x2cv

t, p = paired_ttest_5x2cv(estimator1=c1,
                          estimator2=c3,
                          X=X, y=y,
                          random_seed=25)

print('t statistic: %.3f' % t)
print('p value: %.3f' % p)
```

# Test McNemar

```python
import numpy as np
from mlxtend.evaluate import mcnemar_table
from mlxtend.evaluate import mcnemar

# Tabla de McNemar
tb = mcnemar_table(y_target=y_target, 
                   y_model1=y_model1, 
                   y_model2=y_model2)

print(tb)


# Clculo de valor de Xi
chi2, p = mcnemar(ary=tb, exact=True, corrected=True)
print('chi-squared:', chi2)
print('p-value:', p)
```
